```python
import pandas as pd
import numpy as np
import warnings
import datetime
warnings.filterwarnings('ignore') # no more 'useless' red warnings
```

```python
df1 = pd.read_csv('../files_for_lab/csv_files/marketing_customer_analysis.csv')
```

```python
df1.shape
```

```python
df1.head()
```

```python
standard_columns = [df1.columns[i].lower() for i in range(len(df1.columns))]
standard_columns = [col_name.lower().replace(' ', '_') for col_name in df1.columns]
```

```python
df1.columns = standard_columns
```

```python
df = df1.copy()
```

```python
#Numerical columns

data_num1 = df.select_dtypes(include = np.number)
data_num1
```

```python
#Categorical columns

data_cat = df.select_dtypes(include = np.object)
data_cat
```

```python
#Filling NAN with mean for numeric columns that contain NANs
data_num1['months_since_last_claim'] = data_num1['months_since_last_claim'].fillna((data_num1['months_since_last_claim'].mean())) #inplace=True)
data_num1['number_of_open_complaints'] = data_num1['number_of_open_complaints'].fillna((data_num1['number_of_open_complaints'].quantile(0.5)))# inplace=True)
```

```python
#Filling NAN with empty strings for categorical columns that contain NANs
data_cat['state'] = data_cat['state'].apply(lambda x:str(x).replace('NAN', ''))
```

```python
data_cat['response'] = data_cat['response'].apply(lambda x:str(x).replace('NAN', ''))
```

```python
data_cat['vehicle_class'] = data_cat['vehicle_class'].apply(lambda x:str(x).replace('NAN', ''))
```

```python
data_cat['vehicle_size'] = data_cat['vehicle_size'].apply(lambda x:str(x).replace('NAN', ''))
```

```python
data_cat['vehicle_type'] = data_cat['vehicle_type'].apply(lambda x:str(x).replace('NAN', ''))
```

```python
data_cat[['effective_to_date']]
```

```python
data_cat['effective_to_date'] = pd.to_datetime(data_cat['effective_to_date'], errors='coerce')
data_cat[['effective_to_date']]
```

```python
data_cat['new_col_month'] = pd.DatetimeIndex(data_cat['effective_to_date']).month
data_cat
```
